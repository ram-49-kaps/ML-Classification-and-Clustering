```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("lab2 data.csv")
df

df.head()
```

```python
salary_columns = ['Min_salary', 'Max_Salary', 'Median_Salary', 'Packages']

def clean_salary(value):
    if pd.isna(value):
        return np.nan
    value = str(value).lower().replace('lpa', '').replace(' ', '').strip()
    if 'cr' in value:
        try:
            return float(value.replace('cr', '')) * 100
        except:
            return np.nan
    try:
        num = float(value)
        if num > 1000:
            return round(num / 1000)
        else:
            return num
    except:
        return np.nan

    try:
        return float(value)
    except:
        return np.nan


for col in salary_columns:
   df[col] = df[col].apply(clean_salary)
df.head(30)
```

```python
percent_columns = ['CIA' , 'GPA' , 'Attendence']

def clean_percentage(val):
    if pd.isna(val):
        return np.nan
    val = str(val).replace('%', '').strip()
    try:
        return float(val)
    except:
        return np.nan

for col in percent_columns:
    df[col] = df[col].apply(clean_percentage)

df[percent_columns].head()
```

```python
# Numeric columns: fill with median
numeric_cols = ['Min_salary', 'Max_Salary', 'Median_Salary', 'Packages']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Percentage/Score columns: fill with mean
df['CIA'] = df['CIA'].fillna(df['CIA'].mean())
df['GPA'] = df['GPA'].fillna(df['GPA'].mean())
df['Attendence'] = df['Attendence'].fillna(df['Attendence'].mean())

# Rating: fill with mode
df['Reviwes'] = df['Reviwes'].fillna(df['Reviwes'].mode()[0])

# Categorical: fill with 'unknown'
categorical_cols = ['Highest_Pay_Comp', 'Internships_Interests', 'Job_Role']
df[categorical_cols] = df[categorical_cols].fillna('unknown')

df = df.round(2)
df.head(30)
```

```python
df.to_csv("lab2 data.csv" , index=False)
```

```python
text_columns = [
    'Job_Role',
    'Highest_Pay_Comp',
    'Internships_Interests'
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.lower()
df
```

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform


data = {
    'Feature1': [1, 2, 3],
    'Feature2': [4, 6, 8]
}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
print("Dataset:")
print(df)

euclidean_dist = squareform(pdist(df, metric='euclidean'))
euclidean_df = pd.DataFrame(euclidean_dist, index=df.index, columns=df.index)
print("\nEuclidean Distance Matrix:")
print(euclidean_df)

manhattan_dist = squareform(pdist(df, metric='cityblock'))  # cityblock = Manhattan
manhattan_df = pd.DataFrame(manhattan_dist, index=df.index, columns=df.index)
print("\nManhattan Distance Matrix:")
print(manhattan_df)
```
